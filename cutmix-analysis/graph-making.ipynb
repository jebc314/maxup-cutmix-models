```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
resnet = []
with open("ResNet50.txt") as f:
    for line in f.readlines():
        if "Valid Accuracy" in line:
            colon_index = line.index(":")
            resnet.append(float(line[colon_index+1:])*100)

resnet_cutmix = []
with open("ResNet50+CutMix.txt") as f:
    for line in f.readlines():
        if "Valid Accuracy" in line:
            colon_index = line.index(":")
            resnet_cutmix.append(float(line[colon_index+1:])*100)

resnet_cutmix_maxup = []
with open("ResNet50+CutMix+MaxUp(m=4).txt") as f:
    for line in f.readlines():
        if "Valid Accuracy" in line:
            colon_index = line.index(":")
            resnet_cutmix_maxup.append(float(line[colon_index+1:])*100)

resnet_cutmix_maxup_fine = []
with open("ResNet50+CutMix+MaxUp(m=4)(fine-tuning).txt") as f:
    for line in f.readlines():
        if "Valid Accuracy" in line:
            colon_index = line.index(":")
            resnet_cutmix_maxup_fine.append(float(line[colon_index+1:])*100)
```

```python
plt.plot(resnet, label="ResNet-50")
plt.plot(resnet_cutmix, label="ResNet-50 with CutMix")
plt.plot(resnet_cutmix_maxup, label="ResNet-50 with CutMix and MaxUp")
plt.plot(resnet_cutmix_maxup_fine, label="ResNet-50 with CutMix and MaxUp (Fine Tuned)")
plt.legend()
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Comparison of ResNet-50 with Different Augmentation Techniques")
```

```python
print(f"ResNet-50: {max(resnet)}")
print(f"ResNet-50 with CutMix: {max(resnet_cutmix)}")
print(f"ResNet-50 with CutMix and MaxUp: {max(resnet_cutmix_maxup)}")
print(f"ResNet-50 with CutMix and MaxUp (Fine Tuned): {max(resnet_cutmix_maxup_fine)}")
```
